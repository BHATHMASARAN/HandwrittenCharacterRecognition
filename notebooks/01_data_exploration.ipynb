# Data Exploration — Handwritten Character Recognition
Visualize MNIST/EMNIST samples, class distribution, and augmentation effects.

```python
import sys
sys.path.insert(0, '..')

import matplotlib.pyplot as plt
from src.data.dataset import get_dataloaders

train_loader, val_loader, test_loader, label_info = get_dataloaders()
print('Classes:', label_info['num_classes'])
print('Train batches:', len(train_loader))
```

```python
images, labels = next(iter(train_loader))

fig, axes = plt.subplots(4, 8, figsize=(14, 7))
for i, ax in enumerate(axes.flat):
    img = images[i][0].numpy()
    ax.imshow(img, cmap='gray')
    ax.set_title(label_info['idx_to_char'][labels[i].item()], fontsize=10)
    ax.axis('off')
plt.tight_layout()
plt.show()
```

```python
import numpy as np
all_labels = []
for _, labels in train_loader:
    all_labels.extend(labels.tolist())

unique, counts = np.unique(all_labels, return_counts=True)
plt.figure(figsize=(14, 4))
plt.bar([label_info['idx_to_char'][u] for u in unique], counts, color='#00c2d1')
plt.title('Class Distribution')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
